```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv('data/ProfileData/input_frame_20240312094407_8.csv')
all_data.head()
```

```python
def plot_laser(plt, df, id):
    datas = df[df['laser'] == id]
    plt.scatter(datas['x'], datas['y'], label="laser-" + str(id))
```

```python
#%matplotlib
plt.figure(figsize=(20, 20))
# 数据图
plt.axis('equal') # x和y轴比例一样
plot_laser(plt, all_data, 0)
plot_laser(plt, all_data, 1)
plot_laser(plt, all_data, 2)
plot_laser(plt, all_data, 10)
plot_laser(plt, all_data, 11)
plot_laser(plt, all_data, 12)
plt.legend()
```
